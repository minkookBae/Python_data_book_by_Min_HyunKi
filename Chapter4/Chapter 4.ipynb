```python
import time
from tqdm import tqdm_notebook

for gu in tqdm_notebook(gu_names):
    element = driver.find_element_by_id("SIGUNGU_NM0")
    element.send_keys(gu)
    
    time.sleep(2)
    xpath = """//*[@id="searRgSelect"]"""
    element_sel_gu = driver.find_element_by_xpath(xpath).click()
    
    time.sleep(1)
    xpath2 = """//*[@id="glopopd_excel"]"""
    element_get_excel = driver.find_element_by_xpath(xpath2).click()
    
    time.sleep(1)
```

```python
import pandas as pd
from glob import glob
```

```python
station_files = glob("../data/지역*.xls")
```

```python
station_files
```

```python
tmp_raw = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)
    


station_raw = pd.concat(tmp_raw)
```

```python
station_raw
```

```python
station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({"Oil_store" : station_raw["상호"],"주소":station_raw["주소"],"가격":station_raw["휘발유"],"셀프":station_raw["셀프여부"],"상표":station_raw["상표"]})
```

```python
stations.head()
```

```python
stations["구"] = [eachAddress.split()[1] for eachAddress in stations["주소"]]
```

```python
stations.head()
```

```python
stations["구"].unique()
```

```python
stations[stations["구"]=="서울특별시"]
```

```python
stations.loc[stations["구"]=="서울특별시", "구"] = "성동구"
stations["구"].unique()
```

```python
stations[stations["구"]=="특별시"]
```

```python
stations.loc[stations["구"]=="특별시", "구"] = "도봉구"
stations["구"].unique()
```

```python
stations = stations[stations["가격"]!="-"]
```

```python
stations.head()
```

```python
stations["가격"] = [float(value) for value in stations["가격"]]
```

```python
stations.reset_index(inplace=True)
del stations["index"]
```

```python
oil_price_top10 = stations.sort_values(by="가격", ascending=False).head(10)
```

```python
oil_price_top10
```

```python
oil_price_bottom10 = stations.sort_values(by="가격",ascending=True).head(10)
oil_price_bottom10
```

```python
gamp_key = "----------------------------" ## 개인 API 키

gmaps = googlemaps.Client(key=gamp_key)
```

```python
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(oil_price_top10.index):
    try:
        tmp_add = str(oil_price_top10["주소"][n]).split("(")[0]
        tmp_map = gmaps.geocode(tmp_add)
        tmp_loc = tmp_map[0].get("geometry")
        lat.append(tmp_loc["location"]["lat"])
        lng.append(tmp_loc["location"]["lng"])
    
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Error")
        
oil_price_top10["lat"] = lat
oil_price_top10["lng"] = lng
oil_price_top10
```

```python
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(oil_price_bottom10.index):
    try:
        tmp_add = str(oil_price_bottom10["주소"][n]).split("(")[0]
        tmp_map = gmaps.geocode(tmp_add)
        tmp_loc = tmp_map[0].get("geometry")
        lat.append(tmp_loc["location"]["lat"])
        lng.append(tmp_loc["location"]["lng"])
    
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Error")
        
oil_price_bottom10["lat"] = lat
oil_price_bottom10["lng"] = lng
oil_price_bottom10
```

```python
map = folium.Map(location=[37.55,126.972], zoom_start = 10.5)

for n in oil_price_top10.index:
    if pd.notnull(oil_price_top10["lat"][n]):
        folium.CircleMarker([oil_price_top10["lat"][n], oil_price_top10["lng"][n]], radius = 15 ,color = "#CD3181", fill_color = "#CD3181").add_to(map)

for n in oil_price_bottom10.index:
    if pd.notnull(oil_price_bottom10["lat"][n]):
        folium.CircleMarker([oil_price_bottom10["lat"][n], oil_price_bottom10["lng"][n]], radius = 15 ,color = "#3186cc", fill_color = "#3186cc").add_to(map)        
```
